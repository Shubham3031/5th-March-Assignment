```python
# 2.Ans.

In Bokeh, glyphs are the basic visual building blocks of a plot, such as 
circles, squares, lines, and text. They are created using the figure() method and 
added to a plot using glyph methods such as circle(), square(), line(), and text(). 
Each glyph method takes data to plot and glyph parameters like size, color, and alpha, among others.
```

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

plot = figure(title='My Plot', x_axis_label='X-Axis', y_axis_label='Y-Axis')

plot.circle(x, y, size=10, color='red', alpha=0.5)
plot.square(x, y, size=10, color='blue', alpha=0.5)
plot.line(x, y, line_width=2, color='green')
plot.text(x, y, text=['A', 'B', 'C', 'D', 'E'], text_font_size='12pt')

output_file('my_plot.html')

show(plot)
```

```python
# 3.Ans.

from bokeh.plotting import figure, show
from bokeh.models import Legend, LegendItem
from bokeh.io import output_notebook

output_notebook()

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="My Plot", x_axis_label='X Axis', y_axis_label='Y Axis')

p.circle(x, y, size=10, color='navy', alpha=0.5, legend_label='My Scatter Plot')
p.title.text_color = "red"
p.title.text_font_size = "20px"
p.xaxis.axis_label_text_color = "green"
p.yaxis.axis_label_text_color = "green"
p.legend.label_text_color = "orange"
p.legend.label_text_font_size = "14pt"

show(p)

```

```python
# 4.Ans.

A Bokeh server is a Python server that allows you to create and deploy 
interactive Bokeh plots and applications that can be updated in real-time. 
Bokeh server enables you to build highly interactive and customizable data 
visualization applications by combining the power of Bokeh with the flexibility
and scalability of Python on the server-side.

To use Bokeh server to create interactive plots that can be updated in real-time, you need to follow these steps:

Create a Bokeh figure and add the necessary glyphs and data sources to it.
Define a Python function that defines the logic of your interactive plot or application. 
This function should take a Bokeh Document as input and modify the Bokeh figure based on 
user interactions or data changes.
Create a Bokeh Server instance and define a Bokeh Application that uses your Python
function to modify the Bokeh figure.
Start the Bokeh server and run your Bokeh application.
```
